```python
# Merge the 'sales' and 'item_master' dataframes using a left join on the 'Item Number' column
# and assign the resulting dataframe to a new variable called 'sales_item_master'

sales_item_master = pd.merge(sales, item_master, how='left', left_on='Item Number', right_on='Item Number')

# Display the first 5 rows of the 'sales_item_master' dataframe
sales_item_master.head()
```

```python
# find common columns between the dataframes
set(customers.columns).intersection(set(sales.columns))
```

```python
# Merge the 'sales' and 'customers' dataframes using a left join on the 'Customer Number' column
# and assign the resulting dataframe to a new variable called 'sales_customers'

sales_customers = pd.merge(sales, customers, how='left', left_on='Customer Number', right_on='Customer Number')

# Display the first 5 rows of the 'sales_customers' dataframe
sales_customers.head()
```

```python
# find common columns between the dataframes
set(cities.columns).intersection(set(customers.columns))
```

```python
# Merge the 'customers' and 'cities' dataframes using a left join on the 'City Code' column
# and assign the resulting dataframe to a new variable called 'customers_cities'
customers_cities = pd.merge(customers, cities, how='left', left_on='City Code', right_on='City Code')

# Display the first 5 rows of the 'customers_cities' dataframe
customers_cities.head()
```

```python
# Display the first 5 rows of the 'sales_item_master' dataframe
sales_item_master.head()
```

```python
#Group the sales_item_master data by product line 
#and calculate count, mean, min, max, and sum for the 'Sales' column

#identify which Product Line have the highest total sales

sales_item_master.groupby('Product Line')['Sales'].agg(['count', 'mean', 'min', 'max', 'sum']).sort_values('sum', ascending = False)
```

```python
#Group the sales_item_master data by Product Group 
#and calculate count, mean, min, max, and sum for the 'Sales' column

#identify which Product Group have the highest total sales

sales_item_master.groupby('Product Group')['Sales'].agg(['count', 'mean', 'min', 'max', 'sum']).sort_values('sum', ascending = False)
```

```python
#Group the sales_item_master data by Item Number
#and calculate count, mean, min, max, and sum for the 'Sales' column

#identify which items have the highest total sales

sales_item_master.groupby('Item Number')['Sales'].agg(['count', 'mean', 'min', 'max', 'sum']).sort_values('sum', ascending = False)
```

```python
#Group the sales_item_master data by Customer Number
#and calculate count, mean, min, max, and sum for the 'Sales' column

#identify which Customer have the highest total sales
sales_item_master.groupby('Customer Number')['Sales'].agg(['count', 'mean', 'min', 'max', 'sum']).sort_values('sum', ascending = False)
```

```python
# convert the 'Date' column to datetime format
sales_item_master['Date'] = pd.to_datetime(sales_item_master['Date'])

#extract year, month, and day from the 'Date' column in the sales_item_master 
sales_item_master['year'] = sales_item_master['Date'].dt.year
sales_item_master['month'] = sales_item_master['Date'].dt.month
sales_item_master['day'] = sales_item_master['Date'].dt.day


#Group the sales_item_master data by year
#and calculate count, mean, min, max, and sum for the 'Sales' column

#identify which year have the highest total sales
sales_item_master.groupby('year')['Sales'].agg(['count', 'mean', 'min', 'max', 'sum']).sort_values('sum', ascending = False)
```

```python
#identify which month have the highest total sales
sales_item_master.groupby(['year', 'month'])['Sales'].agg(['count', 'mean', 'min', 'max', 'sum']).sort_values('sum', ascending = False)

```

```python
#identify which day have the highest total sales
sales_item_master.groupby(['year', 'month', 'day'])['Sales'].agg(['count', 'mean', 'min', 'max', 'sum']).sort_values('sum', ascending = False)
```

```python
#identify Sales trends over time

# group sales data by year and month, and plot as a bar chart
sales_item_master.groupby(['year', 'month'])['Sales'].agg('sum').unstack().plot(kind='bar')

# add title to the plot
plt.title('Monthly Sales Report (January 2012 - June 2014)')
```
